```
#This Python script returns a dataframe containing a setlist with the most played songs by an artist in the last 20 shows.

#Import Modules
import requests
import csv
import sys
import math
import json
import pandas as pd
import numpy as np


# You need to apply for a setlist.fm API key to use the data, it's free for non-commercial projects.
# You can get it here: http://api.setlist.fm/docs/index.html. 

# Call Setlist.fm API
url = 'https://api.setlist.fm/rest/1.0/artist/' + '65f4f0c5-ef9e-490c-aee3-909e7ae6b2ab' + '/setlists?p=1'
#65f4f0c5-ef9e-490c-aee3-909e7ae6b2ab is the Metallica Musicbrainz MBID code, you can search for other artist's codes here https://musicbrainz.org/doc/MusicBrainz_Database

headers = {'Accept': 'application/json', 'x-api-key': 'INSERT KEY HERE'}
r = requests.get(url, headers=headers)
# Get .json Data
data=r.json()

# Total Number of Pages needed to load
totalshows = int(data['total'])
pages = int(math.ceil(totalshows/20))
show_maindf=list()

#For test purposes the number of pages is set to 1 but the range can be adjusted to range(1, pages) in order to search in all the available pages
for page in range(1, 2):
  url = 'https://api.setlist.fm/rest/1.0/artist/' + '65f4f0c5-ef9e-490c-aee3-909e7ae6b2ab' + '/setlists?p=' + str(page)   
  headers = {'Accept': 'application/json', 'x-api-key': 'INSERT KEY HERE'}
  r = requests.get(url, headers=headers)
  data=r.json()
  
  for k in range(0,len(data['setlist'])):
    # Check the date of the show, the Tour name, the city and the country)
    date = data['setlist'][k]['eventDate']
    tour = data['setlist'][k]['tour']['name']
    city = data['setlist'][k]['venue']['city']['name']
    country = data['setlist'][k]['venue']['city']['country']['name']
    header=city+', '+country+', '+date
    
    main_show=list()
    main_show.append(header)
          
    try:
      for n_songs in range (0, len(data['setlist'][k]['sets']['set'][0]['song'])):
        main_show.append(data['setlist'][k]['sets']['set'][0]['song'][n_songs]['name'])

      for n_encore in range (0, (len(data['setlist'][k]['sets']['set'][1]['song']))):
        main_show.append(data['setlist'][k]['sets']['set'][1]['song'][n_encore]['name'])
         
      show_maindf.append(main_show)
    except:
      continue    
    
    
```

```
show_maindf=np.array(show_maindf)
show_maindf=show_maindf.T
dftest=pd.DataFrame(show_maindf)
dftest.columns = dftest.iloc[0]
dftest=dftest.drop([0])
dftest
```
